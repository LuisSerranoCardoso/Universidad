```python
def process_text(text, word_to_replace):
    """Reemplaza las palabras indicadas en el texto por asteriscos y cuenta cu√°ntas han sido reemplazadas."""
    processed_text = text.lower().strip()

    replace_count = 0
    for word in word_to_replace:
        word_lower = word.lower()
        if word_lower in processed_text:
            occurrences = processed_text.count(word_lower)
            replace_count += occurrences
            processed_text = processed_text.replace(word_lower, '*' * len(word_lower))

    return processed_text, replace_count


def main():
    text = input("Introduce el texto a procesar: ")

    palabras_a_reemplazar = input("Introduce la lista de palabras a reemplazar (separadas por espacios o comas): ")
    word_to_replace = [w.strip() for w in palabras_a_reemplazar.replace(',',' ').split() if w.strip()]
    processed, count = process_text(text, word_to_replace)   
    print(f"Texto procesado: {processed}") 
    print(f"N√∫mero de palabras reemplazadas: {count}")


if __name__ == "__main__":
    main()
```

```python
from math import isqrt
from typing import List

def is_prime(n: int) -> bool:
    """Determina si un n√∫mero entero es primo."""
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True 

def prime_list(limit: int) -> List[int]:
    """Devuelve una lista de n√∫meros primos hasta el l√≠mite dado."""
    if limit < 2:
        return []
    return [n for n in range(2, limit + 1) if is_prime(n)]

def check_palindrome(primes: List[int]) -> List[int]:
    """Filtra y devuelve los n√∫meros primos que son pal√≠ndromos."""
    return [p for p in primes if str(p) == str(p)[::-1]]

def categorize_primes(prime: int) -> str:
    """Clasifica un n√∫mero primo como peque√±o, mediano o grande seg√∫n su valor."""
    if prime < 10:
        return "peque√±os"
    elif prime < 100:
        return "medianos"
    return "grandes"

def main():
    limit = int(input("Introduce el l√≠mite superior para generar n√∫meros primos: "))
    primes = prime_list(limit)
    print(f"N√∫meros primos hasta {limit}: {primes}")

    palindromic_primes = check_palindrome(primes)
    print(f"N√∫meros primos pal√≠ndromos: {palindromic_primes}")

    categorized_primes = {"peque√±os": [], "medianos": [], "grandes": []}
    for prime in primes:
        category = categorize_primes(prime)
        categorized_primes[category].append(prime)

    for category, nums in categorized_primes.items():
        print(f"N√∫meros primos {category}: {nums}")
    
if __name__ == "__main__":
    main()
```

```python
def leer_lista() -> list[int]:
    """Lee una lista de n√∫meros enteros desde la entrada del usuario."""
    while True:
        entrada = input("Introduce una lista de n√∫meros enteros separados por espacios: ")
        return [int(x) for x in entrada.replace(',', ' ').split() if x.strip()]

def suma_total(numeros: list[int]) -> int:
    """Devuelve la suma total de los n√∫meros de la lista."""
    return sum(numeros)

def suma_pos_y_neg(numeros: list[int]) -> (int, int):
    """Devuelve la suma de los n√∫meros positivos y la suma de los negativos."""
    suma_pos = sum(x for x in numeros if x > 0)
    suma_neg = sum(x for x in numeros if x < 0)
    return suma_pos, suma_neg     

def maximo_minimo(numeros: list[int]) -> (int, int):
    """Devuelve el n√∫mero m√°ximo y el m√≠nimo de la lista."""
    return max(numeros), min(numeros)

def son_unicos(numeros: list[int]) -> bool:
    """Comprueba si todos los n√∫meros de la lista son √∫nicos."""
    return len(numeros) == len(set(numeros))
    
def comparar_singos(numeros: list[int]) -> bool:
    """Compara la cantidad de n√∫meros positivos y negativos."""
    positivos = sum(1 for x in numeros if x > 0)
    negativos = sum(1 for x in numeros if x < 0)
    return ("M√°s positivos" if positivos > negativos else
            "M√°s negativos" if negativos > positivos else
            "Iguales"
    )

def mostrar_menu():
    """Muestra el men√∫ de opciones en pantalla."""
    print("\nMen√∫ de opciones:")
    print("1. Suma total")
    print("2. Suma de positivos y negativos")
    print("3. M√°ximo y m√≠nimo")
    print("4. Comprobar si todos los n√∫meros son √∫nicos")
    print("5. Comparar cantidad de positivos y negativos")
    print("6. Salir")

def main():
    numeros = leer_lista()
    while True:
        print("\nLa lista de n√∫meros es:", numeros)
        mostrar_menu()
        opcion = input("Selecciona una opci√≥n (1-6): ")
        if opcion == '1':
            print(f"Suma total: {suma_total(numeros)}")
        elif opcion == '2':
            suma_pos, suma_neg = suma_pos_y_neg(numeros)
            print(f"Suma de positivos: {suma_pos}, Suma de negativos: {suma_neg}")
        elif opcion == '3':
            max_num, min_num = maximo_minimo(numeros)
            print(f"M√°ximo: {max_num}, M√≠nimo: {min_num}")
        elif opcion == '4':
            print("Todos los n√∫meros son √∫nicos." if son_unicos(numeros) else "Hay n√∫meros repetidos.")
        elif opcion == '5':
            print(comparar_singos(numeros))
        elif opcion == '6':
            print("Saliendo del programa.")
            break
        else:
            print("Opci√≥n no v√°lida. Int√©ntalo de nuevo.")

if __name__ == "__main__":   
    main()
```

Ejercicio 4

```python
def add_task(task: str) -> None:
    """A√±ade una tarea a la lista global de tareas."""
    global tareas
    tareas.append(task)
    print(f"‚úÖ Tarea '{task}' a√±adida.")


def remove_task(task: str) -> None:
    """Elimina una tarea de la lista global si existe."""
    global tareas
    try:
        tareas.remove(task)
        print(f"üóëÔ∏è Tarea '{task}' eliminada.")
    except ValueError:
        print(f"‚ö†Ô∏è Tarea '{task}' no encontrada.")


def list_tasks() -> None:
    """Muestra todas las tareas pendientes."""
    if not tareas:
        print("üì≠ No hay tareas en la lista.")
    else:
        print("Tareas pendientes:")
        for idx, task in enumerate(tareas, 1):
            print(f"{idx}. {task}")


def task_manager() -> None:
    """Gestiona las tareas permitiendo a√±adir, eliminar, modificar y listar."""
    def modify_task():
        """Permite modificar una tarea existente."""
        global tareas
        list_tasks()
        try:
            idx = int(input("Introduce el n√∫mero de la tarea a modificar: ")) - 1
            if 0 <= idx < len(tareas):
                new_task = input("Introduce la nueva descripci√≥n de la tarea: ").strip()
                if new_task:
                    old_task = tareas[idx]
                    tareas[idx] = new_task
                    print(f"‚úèÔ∏è Tarea '{old_task}' modificada a '{new_task}'.")
                else:
                    print("‚ö†Ô∏è La descripci√≥n de la tarea no puede estar vac√≠a.")
            else:
                print("‚ö†Ô∏è N√∫mero de tarea inv√°lido.")
        except ValueError:
            print("‚ö†Ô∏è Entrada inv√°lida. Por favor, introduce un n√∫mero.")
        
    while True:
        print("\nGestor de Tareas")
        print("1. A√±adir tarea")
        print("2. Eliminar tarea")
        print("3. Modificar tarea")
        print("4. Listar tareas")
        print("5. Salir")
        
        opcion = input("Selecciona una opci√≥n (1-5): ").strip()
        
        if opcion == '1':
            task = input("Introduce la descripci√≥n de la tarea: ").strip()
            if task:
                add_task(task)
            else:
                print("‚ö†Ô∏è La descripci√≥n de la tarea no puede estar vac√≠a.")
        elif opcion == '2':
            task = input("Introduce la descripci√≥n de la tarea a eliminar: ").strip()
            if task:
                remove_task(task)
            else:
                print("‚ö†Ô∏è La descripci√≥n de la tarea no puede estar vac√≠a.")
        elif opcion == '3':
            modify_task()
        elif opcion == '4':
            list_tasks()
        elif opcion == '5':
            print("üëã Saliendo del gestor de tareas.")
            break
        else:
            print("‚ö†Ô∏è Opci√≥n no v√°lida. Int√©ntalo de nuevo.")


def main():
    """Inicia el gestor de tareas pendientes."""
    print("üóìÔ∏è Bienvenido al gestor de tareas pendientes")
    task_manager()


if __name__ == "__main__":   
    main()
```

```python
def imprimir_cadenas_por_letra(cadenas):
    """Imprime las cadenas que comienzan con una letra indicada por el usuario."""
    letra = input("Introduce una letra: ")
    print(f"\nCadenas que comienzan con '{letra}':")
    for cadena in cadenas:
        if cadena.lower().startswith(letra.lower()):
            print(cadena)


def contar_cadenas_con_subcadena(cadenas):
    """Cuenta cu√°ntas cadenas contienen una subcadena dada por el usuario."""
    subcadena = input("Introduce una subcadena: ")
    contador = sum(1 for cadena in cadenas if subcadena.lower() in cadena.lower())
    print(f"N√∫mero de cadenas que contienen '{subcadena}': {contador}")


def encontrar_larga_y_corta(cadenas):
    """Muestra la cadena m√°s larga y la m√°s corta de la lista."""
    if not cadenas:
        print("La lista de cadenas est√° vac√≠a.")
        return
    cadena_larga = max(cadenas, key=len)
    cadena_corta = min(cadenas, key=len)
    print(f"Cadena m√°s larga: {cadena_larga} (Longitud: {len(cadena_larga)})")
    print(f"Cadena m√°s corta: {cadena_corta} (Longitud: {len(cadena_corta)})")


def verificar_mismo_objeto(cadenas):
    """Verifica si dos cadenas son el mismo objeto en memoria usando el operador 'is'."""
    print("\nVerificaci√≥n con operador 'is':")
    try:
        i = int(input("Introduce el √≠ndice de la primera cadena: "))
        j = int(input("Introduce el √≠ndice de la segunda cadena: "))
        if cadenas[i] is cadenas[j]:
            print("Las cadenas son el mismo objeto en memoria.")
        else:
            print("Las cadenas no son el mismo objeto en memoria.")
    except (ValueError, IndexError):
        print("√çndice no v√°lido.")


def verificar_longitud_mayor_10(cadenas):
    """Verifica si alguna cadena tiene una longitud mayor a 10 caracteres."""
    for cadena in cadenas:
        if len(cadena) > 10:
            print(f"La cadena '{cadena}' tiene m√°s de 10 caracteres.")
            break
    else:
        print("Ninguna cadena tiene m√°s de 10 caracteres.")


def main():
    cadenas = input("Introduce una lista de cadenas separadas por comas: ").split(',')
    cadenas = [cadena.strip() for cadena in cadenas if cadena.strip()]

    while True:
        print("\nMen√∫ de opciones:")
        print("1. Imprimir cadenas que comienzan con una letra espec√≠fica")
        print("2. Contar cadenas que contienen una subcadena espec√≠fica")
        print("3. Encontrar la cadena m√°s larga y la m√°s corta")
        print("4. Verificar si dos cadenas son el mismo objeto en memoria")
        print("5. Verificar si alguna cadena tiene m√°s de 10 caracteres")
        print("6. Salir")

        opcion = input("Selecciona una opci√≥n (1-6): ")

        if opcion == '1':
            imprimir_cadenas_por_letra(cadenas)
        elif opcion == '2':
            contar_cadenas_con_subcadena(cadenas)
        elif opcion == '3':
            encontrar_larga_y_corta(cadenas)
        elif opcion == '4':
            verificar_mismo_objeto(cadenas)
        elif opcion == '5':
            verificar_longitud_mayor_10(cadenas)
        elif opcion == '6':
            print("Saliendo del programa.")
            break
        else:
            print("Opci√≥n no v√°lida. Int√©ntalo de nuevo.")


if __name__ == "__main__":   
    main()

            
```

```python
import re
from datetime import datetime


def calcular_edad(fecha_nacimiento: str, fecha_consulta: str) -> int:
    """Calcula la edad en a√±os en la fecha de consulta dada la fecha de nacimiento."""
    nac = datetime.strptime(fecha_nacimiento, "%d/%m/%Y")
    consulta = datetime.strptime(fecha_consulta, "%d/%m/%Y")
    edad = consulta.year - nac.year
    if (consulta.month, consulta.day) < (nac.month, nac.day):
        edad -= 1
    return edad


def procesar_informe(informe: str) -> str:
    """Procesa el informe sustituyendo nombre por 'PACIENTE', nacimiento por edad y la l√≠nea del m√©dico por 'FECHA MEDICO'."""
    fecha_pattern = re.compile(r'\b\d{1,2}/\d{1,2}/\d{4}\b')
    nombre_pattern = re.compile(r'^Nombre:.*')
    medico_pattern = re.compile(r'.*(?:Dr\.|Dra\.).*')
    nacimiento_pattern = re.compile(r'^Fecha de nacimiento:\s*(\d{1,2}/\d{1,2}/\d{4})')
    lineas = informe.splitlines()
    fecha_nacimiento = None
    fecha_consulta = None

    for linea in lineas:
        m_nac = nacimiento_pattern.search(linea)
        if m_nac:
            fecha_nacimiento = m_nac.group(1)
        else:
            m_f = fecha_pattern.search(linea)
            if m_f:
                fecha_consulta = m_f.group(0)

    resultado = []
    for linea in lineas:
        if nombre_pattern.match(linea):
            resultado.append("Nombre: PACIENTE")

        elif nacimiento_pattern.match(linea):
            if fecha_consulta and fecha_nacimiento:
                edad = calcular_edad(fecha_nacimiento, fecha_consulta)
                resultado.append(f"Edad: {edad}")
            else:
                resultado.append(linea)
        elif medico_pattern.match(linea):
            resultado.append("FECHA MEDICO")
        else:
            resultado.append(linea)

    return "\n".join(resultado)


def main() -> None:
    """Ejecuta el procesamiento del informe"""
    informe = """Informe cl√≠nico de Urgencias
Nombre: Juan P√©rez L√≥pez
G√©nero: Hombre
Fecha de nacimiento: 12/05/1980
Motivo de consulta: dolor abdominal
12/05/2023 Dr. Ram√≠rez
Tratamiento: reposo y analg√©sicos"""

    informe_procesado = procesar_informe(informe)
    print(informe_procesado)


if __name__ == "__main__":
    main()

```
