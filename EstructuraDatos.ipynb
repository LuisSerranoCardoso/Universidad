Clase Pila

```python
class Stack:
    def __init__(self):
      self.items = []

    def push(self, item):
      self.items.append(item)

    def pop(self):
      if not self.is_empty():
         return self.items.pop()
      return None

    def peek (self):
       if not self.is_empty():
          return self.items[-1]
       return None

    def is_empty(self):
       return len(self.items)== 0

    def size(self):
        return len(self.items)

    def __str__(self):
      return f"Mi pila es {self.items[::-1]}"

stack = Stack()
stack.push(1)
stack.push(2)
stack.push(3)
print(stack.pop())
print(stack.peek())
print(stack.size())
print(stack.is_empty())
print(stack)
```

Ejercicio Diccionario: estudiantes y notas

```python
#Crear un diccionario vacio

estudiantes = {}

#Bucle para introducir nombre y notas
while True:
    name = input("Introduzca nombre \"done\"  para terminar: ")
    if name.lower() == 'done':
      break

    grade = float(input("Introduzca nota: "))
    estudiantes[name] = grade

#Imprimir los estudiantes y sus notas

for name, grade in estudiantes.items():
    print(f"{name}: {grade}")

#Calcular la media
if estudiantes:
  average = sum(estudiantes.values())/len(estudiantes)
  print(f"La media es {average: .2f}")
else:
  print("No ha introducido estudiantes")


#Calcular la nota más alta
if estudiantes:
   max_nota = max(estudiantes.values())
   print(f"La nota más alta es {max_nota}")
else:
  print("No ha introducido estudiantes")


#Imprimir el nombre del estudiante(s) con la nota más alta
if estudiantes:
  top_estudiante = [name for name, grade in estudiantes.items() if grade == max_nota]
  print(f"El(los) estudiante(s) con la nota más alta es {top_estudiante}")

```

Ejercicio de Gestor de tareas

```python
from enum import Enum, auto

class EstadoTareas(Enum):
    NOT_EMPEZADA = auto()
    IN_PROGRESO = auto()
    COMPLETADA = auto()
    BLOQUEADA = auto()

# Uso de imprimir Tareas
def imprimir_tareas(nombret, estado):
    if estado == EstadoTareas.NOT_EMPEZADA:
        print(f"The task '{nombret}' has not started yet.")
    elif estado == EstadoTareas.IN_PROGRESO:
        print(f"The task '{nombret}' is currently in progress.")
    elif estado == EstadoTareas.COMPLETADA:
        print(f"The task '{nombret}' is completed.")
    elif estado == EstadoTareas.BLOQUEADA:
        print(f"The task '{nombret}' is currently blocked.")

# Probar el ejemplo
imprimir_tareas("Correción exámenes",EstadoTareas.NOT_EMPEZADA)
```
